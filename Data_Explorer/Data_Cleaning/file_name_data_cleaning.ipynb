```python
#df = df.rename(columns={'Unnamed: 0': "ID"})
df = df.drop("Unnamed: 0", axis=1)
```

```python
df['file_path'] = df['file_path'].str.replace(r'^\\\\\?\\', '', regex=True)
df.head()
```

```python

# List of target columns
target_columns = ["pdf", "iam", "ipt", "dwg", "rfa", "sldprt", "sldasm", "step", "idw"]

# Filter: Rows where any of the target_columns is greater than 0
filtered_df = df[df[target_columns].gt(0).any(axis=1)]

# Select only the desired columns
description_columns = ['parent_folder', 'file_name', 'file_path']
filtered_columns = description_columns + target_columns
result_df = filtered_df[filtered_columns]

result_df.head()
```

```python
result_df.info()
```

```python
# Filter for folder
# Folder filtered with narrowed_folders list
# Explicit exclusion of "Fabrication - Obsolete" in excluded_path
# Additional exclusion of any path containing "OUT-OF-DATE" (case-insensitive)

narrowed_folders = ["Fabrication", "Fabrication Ulma", "Prodtech", "Test Reports"]
path_nf = r"H:\CACLA - Caledon ON\Product and Technical"

# Generate full paths to match against
valid_paths = [f"{path_nf}\\{folder}" for folder in narrowed_folders]

# Filter the DataFrame to only include rows with matching start of path
df_filtered = result_df[result_df['file_path'].str.startswith(tuple(valid_paths))]

# Explicitly exclude "Fabrication - Obsolete"
excluded_path = f"{path_nf}\\Fabrication - Obsolete"
df_filtered = df_filtered[~df_filtered['file_path'].str.startswith(excluded_path)]

# Exclude any path that contains "OUT-OF-DATE"
df_filtered = df_filtered[~df_filtered['file_path'].str.contains("OUT-OF-DATE", case=False, na=False)]

# Optional: Display the result
print(df_filtered.head())

```

```python
# Filter out rows where 'parent_folder' contains any obsolete/old keywords (case-insensitive)
pattern = r'(?:^|[^a-zA-Z])(obsolete|old|oldversions)(?:[^a-zA-Z]|$)'
df_cleaned = df_filtered[~df_filtered['parent_folder'].str.contains(pattern, case=False, na=False, regex=True)]
```

```python
# Inverse of df_cleaned → rows that were filtered out
df_removed = df_filtered[df_filtered['parent_folder'].str.contains(r'\b(obsolete|old|oldversions)\b', case=False, na=False)]

# Display count and a preview
print(f"Rows removed: {len(df_removed)}")
#print(df_removed[['parent_folder', 'file_path']].head(10))

df_removed.to_csv("checker_removed.csv", index=False)
```

```python
# Define the keywords we want to flag (case-insensitive)
keywords = ['old', 'obsolete', 'oldversions']

# Create a regex pattern that matches the keyword as a folder name (e.g., \old\, \obsolete\)
pattern = r'\\(' + '|'.join(keywords) + r')\\'

# Filter rows where file_path contains any of the keywords as a full folder
mask = df_cleaned['file_path'].str.contains(pattern, case=False, regex=True, na=False)

# Show only the suspicious paths
df_flagged = df_cleaned[mask]

print(f"Suspicious folders found: {len(df_flagged)}")
print(df_flagged[['file_path']].head())


df_flagged.to_csv("checker.csv", index=False)
```

```python
def classify_main_folder(path):
    norm_path = os.path.normpath(path).lower()
    path_parts = norm_path.split(os.sep)
    for folder in narrowed_folders:
        if folder.lower() in path_parts:
            return folder
    return "Other"

# Apply to DataFrame
df_cleaned['main_folder'] = df_cleaned['file_path'].apply(classify_main_folder)

# Optional: preview grouped counts
print(df_cleaned['main_folder'].value_counts())
```
