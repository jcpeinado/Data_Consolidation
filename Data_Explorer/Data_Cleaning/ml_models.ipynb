```python
# Libraries

import pandas as pd
import matplotlib.pyplot as plt
import os
from rapidfuzz import fuzz
```

```python
# Reading the cleaned data
df = pd.read_csv("Processed_Data.csv")
```

```python
ratios = []
for _, row in df.iterrows():
    ratio = fuzz.ratio(row['item_no'], row['file_name'])
    ratios.append(ratio)

df["fuzzy_ratio"] = (ratios)

df.head()
```

```python
df['match_tag'] = df['fuzzy_ratio'].apply(
    lambda x: 'High' if x >= 90 else 'Medium' if x >= 70 else 'Low'
)

df.head()
```

```python
df_high_cutoff = df[df['fuzzy_ratio'] >= 90]   
df_medium_cutoff = df[df['fuzzy_ratio'] >= 70]

```

```python
## Files exported
df.to_csv("df_ML.csv", index=False)
df_medium_cutoff.to_csv("df_medium_cutoff.csv", index=False)
df_high_cutoff.to_csv("df_high_cutoff.csv", index=False)
```
